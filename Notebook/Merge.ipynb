```python
import pandas as pd

installments_dfs = pd.read_csv('../Data/installments_payments_dfs.csv')
print(installments_dfs.shape)
installments_dfs.head()
```

```python
cash_dfs = pd.read_csv('../Data/POS_CASH_balance_dfs.csv')
print(cash_dfs.shape)
cash_dfs.head()
```

```python
credit_dfs = pd.read_csv('../Data/credit_card_balance_dfs.csv')
print(credit_dfs.shape)
credit_dfs.head()
```

```python
pre_app = pd.read_csv('../Data/previous_application.csv')
print(pre_app.shape)
pre_app.head()
```

```python
pre_app_merged = pd.merge(pre_app, installments_dfs, how='left')
pre_app_merged = pd.merge(pre_app_merged, cash_dfs, how='left')
pre_app_merged = pd.merge(pre_app_merged, credit_dfs, how='left')
pre_app_merged.shape
```

```python
pre_app_merged.to_csv('../Data/pre_app_merged.csv', index=False, encoding='utf-8_sig')
```

```python
import pandas as pd

app_train = pd.read_csv('../Data/application_train.csv')
app_test = pd.read_csv('../Data/application_test.csv')
app_test['TARGET'] = -1
app = pd.concat([app_train, app_test], axis=0)
app.head()
```

```python
app['TARGET'].unique()
```

```python
len(app.SK_ID_CURR.unique())
```

```python
pre_app_merged_dfs = pd.read_csv('../Data/pred_app_merged_dfs.csv')
pre_app_merged_dfs.head()
```

```python
bureau_merged_dfs = pd.read_csv('../Data/bureau_merged_dfs.csv')
bureau_merged_dfs.head()
```

```python
app_merged = pd.merge(app, pre_app_merged_dfs, how='left')
app_merged = pd.merge(app_merged, bureau_merged_dfs, how='left')
app_merged.shape
```
