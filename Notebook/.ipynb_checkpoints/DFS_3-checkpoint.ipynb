```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 22
%matplotlib inline
import seaborn as sns

# Suppress warnings from pandas
import warnings
warnings.filterwarnings('ignore')
```

```python
credit = pd.read_csv('../Data/credit_card_balance.csv').sort_values(['SK_ID_CURR', 'SK_ID_PREV']).reset_index(drop = True)
```

```python
data = credit
data_name = 'credit'
```

```python
data.head()
```

```python
obj_feature = []
int_feature = []
float_feature = []
for col in data.columns:
    if data[col].ftype == 'object:dense':
        obj_feature.append(col)
    elif data[col].ftype == 'int64:dense':
        int_feature.append(col)
    else:
        float_feature.append(col)
```

```python
obj_name = []
for col in obj_feature:
    tmp = data[col].unique()
    tmp_col_name = [data_name+'_'+col+'_'+str(col_) for col_ in tmp]
    obj_name.extend(tmp_col_name)

int_name = []
for col in int_feature:
    tmp_col_name = [data_name+'_'+col+'_min', data_name+'_'+col+'_max']
    int_name.extend(tmp_col_name)
    
float_name = []
for col in float_feature:
    tmp_col_name = [data_name+'_'+col+'_min', data_name+'_'+col+'_max', col+'_mean']
    float_name.extend(tmp_col_name)
```
