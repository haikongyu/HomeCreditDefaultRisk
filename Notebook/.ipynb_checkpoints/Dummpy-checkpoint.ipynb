```python
import pandas as pd


im_data = pd.read_csv('../Data/result_6.csv')
dummpy_data = pd.DataFrame()
for col in im_data.columns:
    if im_data[col].ftype == 'int64:dense' and len(im_data[col].unique()) > 2:
        dummpy = pd.get_dummies(im_data[col])
        col_name = [col+'_'+str(i) for i in dummpy.columns]
        dummpy.columns = col_name
        for col_ in dummpy.columns:
            dummpy_data[col_] = dummpy[col_]
            
    else:
        dummpy_data[col] = im_data[col]
        
im_data = dummpy_data
im_data.to_csv('../Data/result_6_dummpy.csv', encoding='utf-8_sig', index=False)
im_data.head()
```
